# Inference Example

```python
from pycoder.model.inference import CodeInference
import pycoder.config as cfg
```

```python
coder = CodeInference(cfg.MODEL_PATH, cfg.TOKENIZER_PATH, cfg.CONTROL_TOKENS, 1000, True)
```

```python
out = coder("algorithm,general,string,palindrome", "a function to check if two strings are palindrome", "def check_string_palindrome(str1, str2):\n")
```
